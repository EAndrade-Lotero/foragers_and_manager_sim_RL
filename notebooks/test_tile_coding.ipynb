```python
import sys

sys.path.append('../src')

import numpy as np
import pandas as pd
import seaborn as sns

from agents.linearQ import TilesQ
from utils.plot_utils import PlotQApprox2D
from utils.interpreters import gym_interpreter1
from environments.foragers import DiscreteForagersEnv
```

```python
num_actions = 3
env = DiscreteForagersEnv(
    initial_rate=0.5,
    initial_wealth=0.5,
    num_foragers=3,
    num_discrete_actions=num_actions
)
state = env.reset()
```

```python
state_scales = [
    {'min':0.0, 'max':1.0},
    {'min':0.0, 'max':1.0},
]
tiles_parameters = {
    'numDims': 2,
    'numTilings': 8,
    'numTiles': [8, 8],
    'scaleFactors':state_scales,
    'maxSize':4096,
    'alpha':0.1,
    'num_actions': num_actions
}
value_approximator = TilesQ(parameters=tiles_parameters)
```

```python
state_ = gym_interpreter1(state)
```

```python
state = env.reset()
env.render()
state_ = gym_interpreter1(state)
action = 1
print("Tiles activated:", value_approximator.my_tiles(state_, action))
print(f"Q-value for state {state_} and action {action}:", value_approximator.predict(state_, action))
new_state, reward, done, truncated, info = env.step(action)
print("Reward received:", reward)
env.render()
```

```python
value_approximator.learn(state_, action, update=reward)
print(f"New Q-value for state {state_} and action {action}:", value_approximator.predict(state_, action))
```
