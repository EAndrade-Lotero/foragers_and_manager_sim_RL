## Tile coding with the performer

```python
# Define the range of the function
state_scales = [
    {'min':-1.2, 'max':0.6},
    {'min':-0.07, 'max':0.07},
]
tiles_parameters = {'numDims': 2,
                    'numTilings': 8,
                    'numTiles': [8, 8],
                    'scaleFactors':state_scales,
                    'maxSize':4096,
                    'alpha':0.1}
value_approximator = TilesQ(parameters=tiles_parameters)
agent_parameters = {'nA':3,
                    'nS':2,
                    'gamma':1,
                    'epsilon':0,
                    'alpha':0.1,
                    'Q':value_approximator
                    }
env_parameters = {}
perf = Performer(env_name='MountainCar-v0',
                 env_parameters=env_parameters,
                 state_interpreter=gym_interpreter1,
                 agent_name='agentsCS.QLearningCS',
                 agent_parameters=agent_parameters)
```

```python
perf.train(num_rounds=1000, num_episodes=100)
```

```python
dict_acciones = {0:"⬅", 1:"N", 2:"➡"}
q_plot = PlotQApprox2D(
    Q=perf.agent.Q,
    state_scales=state_scales, 
    dict_acciones=dict_acciones,
    mesh_size=20
)
q_plot.best_actions_with_values(max_deci=2)
```

```python
perf.test(num_rounds=500, num_episodes=100)
```
